```python
# Interpolation: Process of esteamating unknown values from the known values
# 1. Liner Interpolation
# 2. Cubic Interpolation
# 3. Polynomial Interpolation
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 3, 5, 7, 9])

plt.scatter(x, y)
```

```python
# 1. Liner interpolation: Shape of data is liner in nature
x_new = np.linspace(1, 5, 10) #create new x values
y_interp = np.interp(x_new, x, y)
```

```python
plt.scatter(x_new, y_interp)
```

```python
# Cubic interpolation

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 8, 27, 64, 125])

plt.scatter(x, y)
```

```python
from scipy.interpolate import interp1d
```

```python
f = interp1d(x, y, kind = 'cubic')
```

```python
f
```

```python
x_new = np.linspace(1, 5, 10)
y_interp = f(x_new)
```

```python
plt.scatter(x_new,y_interp)
```

```python
# Polynomial interpolation
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 1, 25])
p = np.polyfit(x,y,2)
plt.scatter(x,y)
```

```python
x_new = np.linspace(x,y,2)
Y_interp = np.polyval(p, x_new)
```

```python
plt.scatter(x_new,y_interp)
```
