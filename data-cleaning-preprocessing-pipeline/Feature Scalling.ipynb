```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Feature Scalling:
# 1. Standerdisation (mu=0,sigma=1).
# 2. normalisation (min max scaller): Data comes between 0 and 1
# 3. unit vector whse magnitude of 1
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
sns.get_dataset_names()
```

```python
sns.distplot(df['total_bill'])
```

```python
mean = np.mean(df['total_bill'])
std = np.std(df['total_bill'])
```

```python
std_data = []
for i in list(df['total_bill']):
    zscore = (i-mean)/std
    std_data.append(zscore)
```

```python
std_data
```

```python
# Feature scalling is optional and it doesnot change the destribution of data
```

```python
sns.distplot(std_data)
```
