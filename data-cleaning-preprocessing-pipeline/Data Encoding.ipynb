```python
df = pd.DataFrame({
    "Status": ["Singal", "Married","Seperated","Singal", "Married"]
})
```

```python
df
```

```python
encoded_data  = encoder.fit_transform(df[['Status']]).toarray()
encoded_data 
```

```python
encoded_columns  = encoder.get_feature_names_out()
encoded
```

```python
df_encoded = pd.DataFrame(encoded_data, columns=encoded_columns)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({
    "Qualifications": ["HS" ,"GR", "PG", "PHD", "GR", "PG"]
})
df
```

```python
encoder  = OrdinalEncoder(categories = [["HS", "GR", "PG", "PHD"]])
```

```python
encoder.fit_transform(df[["Qualifications"]])
```

```python
encoder.transform([["PG"]])
```

```python
# Target guided ordinal encoded
# relationship with target variable
# when lots of unique categories there
# repalce the category with mean and median of respective groups
```

```python
df = pd.DataFrame({
    'time': ['lunch', 'breackfast', 'dinner', 'lunch', 'breakfast', 'dinner', 'lunch', 'breakfast', 'dinner'],
    'total_bill': [12, 130, 90, 125, 150, 190, 160, 180, 189]
})
df
```

```python
mean_price = df.groupby('time')['total_bill'].mean().to_dict()
mean_price
```

```python
df['time_encoded'] = df['time'].map(mean_price)
```
